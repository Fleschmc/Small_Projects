```python
import pandas as pd
import numpy as np
import time
import itertools
import math
from collections import Counter
pd.set_option('display.max_colwidth', None)
```

```python
def cloning(clist, numr = 4, gen = False, depth = 10, ncross_breed = 6):
    start = time.time()
    genes = {'w' : 0.9,  'x' : 0.9,  'g' : 0.5, 'h' : 0.5, 'y' : 0.5} # How Rust weights genes when calculating dominance
    zero =  {'w' : 0,    'x' : 0,    'g' : 0,   'h' : 0,   'y' : 0}   # Used for calculating dominance
    score = {'w' : -0.5, 'x' : -0.5, 'g' : 1,   'h' : 0.5, 'y' : 1}   # My scoring dictionary
    combs, calculated, master = [], 0, []
    def do_stuff():
        do_stuff.calculated = 0
        for comb in combs:
            icounter = Counter(comb)
            if any([value > (0.5 * len(comb)) for value in list(icounter.values())]): # Removing combs that have >50% clones the same
                pass
            else:
                do_stuff.calculated += 1
                clones = list(comb).copy()
                clones_weighted = [list(zip([genes[gene] for gene in list(clone)], list(clone))) for clone in clones]
                inter = []
                for num in range(6): # six because there are six genes per clone
                    zero_copy = zero.copy()
                    for i in clones_weighted:
                        zero_copy[i[num][1]] += i[num][0] # Calculating which gene is dominant for each gene slot
                    inter.append(zero_copy)
                put, problems, solutions = [max(_, key = _.get) for _ in inter], [], [] # Creating the gene from the scoring scheme / not perfect / max() doesn't account for the 50/50 split
                for i, x in enumerate(inter):
                    duptest, ip = [value == max(x.values()) for value in list(x.values())], [] # duptest tests each gene column for a 50/50 shot
                    if sum(duptest) > 1:
                        for j, y in enumerate(duptest):
                            if y == True:
                                ip.append((i, list(x.keys())[j]))
                    if len(ip) > 0:
                        problems.append(ip)

                if len(problems) > 0:    
                    for x in list(itertools.product(*problems)):
                        for y in x:
                            tput = put.copy()
                            del tput[y[0]]
                            tput.insert(y[0], y[1])
                            solutions.append(tput)
                    solutions = [tuple(_) for _ in solutions]
                    for product in set(solutions):            
                        data_dict = {}
                        data_dict['product']    = ''.join(product)
                        data_dict['score']      = sum([score[p] for p in product])
                        data_dict['clones']     = clones
                        data_dict['num_clones'] = len(clones)
                        master.append(data_dict)
                else:
                    data_dict = {}
                    data_dict['product']    = ''.join(put)
                    data_dict['score']      = sum([score[p] for p in put])
                    data_dict['clones']     = clones
                    data_dict['num_clones'] = len(clones)
                    master.append(data_dict)

        dfd = pd.DataFrame(master)
        dfd['n/o'] = np.where(dfd['product'].isin(clist), 'old', 'new')
        return dfd.sort_values(['score', 'n/o', 'num_clones'], ascending = (False, True, True))[['product', 'score', 'n/o', 'clones']].drop_duplicates(subset=['product'], keep = 'first')
        
    if gen == True:
        gclist, removed = [], [] # Will hold the clones and how they were made from each generation.
        while gen == True:
            
            cstart = time.time()
            combs  = [itertools.combinations_with_replacement(clist, _) for _ in range(1, ncross_breed + 1)]
            combs  = [item for sublist in combs for item in sublist]
            if len(gclist) > 0:
                combs = set([tuple(_) for _ in pd.DataFrame(master)['clones']]) ^ set(combs)  # Removing all the clones already calculatd
            cend   = time.time()
            dstart = time.time()
            df     = do_stuff()
            dend   = time.time()
            
            aclist = df[(df['n/o'] == 'new') & (df['score'] >= max(df['score']) - 3 / max(df['score']))]
            if len(gclist) > 0:
                aclist = aclist[(-aclist['product'].isin(removed)) & (-aclist['product'].isin([_[0] for _ in gclist[-1]]))]
                # Line above ensures we aren't adding in clones that have already been removed and also giving clones a 1 generation grace period before being removed 
            if len(aclist) >= 1:
                if len(aclist) > depth:
                    gclist.append(list(zip(aclist['product'][:depth], aclist['clones'][:depth], aclist['score'][:depth])))
                    clist = clist + aclist['product'].tolist()[:depth]
                else:
                    gclist.append(list(zip(aclist['product'], aclist['clones'], aclist['score'])))
                    clist = clist + aclist['product'].tolist()
                    
                lenclist = len(clist)
                if len(gclist) > 0:
                    cdict = {}
                    for c in clist:
                        cdict[c] = 0
                        for m in aclist['clones']:
                            if c in m:
                                cdict[c] += 1
                        if cdict[c] < 0.05 * len(aclist): # removing clones that aren't in at least 5% of the new clones
                            clist.remove(c)
                            removed.append(c)
                            
                print(f"{len(gclist)}\ncombs: {round(cend - cstart, 2)} \ndo_stuff(): {round(dend - dstart, 2)} \nlen(combs):{do_stuff.calculated} {len(combs) - do_stuff.calculated} \nlen(aclist) {len(aclist)} len(clist) {len(clist)} removed {lenclist - len(clist)}")
            else:
                end = time.time()
                tt = end - start
                print(f"Code took {int(tt //  60)} minutes and {int(tt % 60)} seconds")
                return df[:numr], gclist 
    else:
        combs = [list(itertools.combinations_with_replacement(clist, _)) for _ in range(1, ncross_breed + 1)]
        combs = [item for sublist in combs for item in sublist]
        df = do_stuff()
        end = time.time()
        tt = end - start
        print(f"Code took {int(tt //  60)} minutes and {int(tt % 60)} seconds")
        return df[:numr]

test = cloning('gyywgx ygwwhg hhgyhx whhwgh wygwgg whyhhh hgggyh xygygh whhwhh yyywyh xghyyh xygyyg hygyyg yygwhx'.split(' '), 5, True, 5, 5)
test
```
