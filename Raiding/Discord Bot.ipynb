```python
import requests
from bs4 import BeautifulSoup
from fuzzywuzzy import fuzz
import pandas as pd
import numpy as np
import json
```

```python
# soup to find build links
soup = BeautifulSoup(requests.get('https://rustlabs.com/group=building-blocks').content, 'html.parser')
```

```python
build_links = ['https://rustlabs.com' + soup.find_all('td', class_ = 'left')[_].next['href'] + '#tab=destroyed-by;filter=0,0,1,0,0,0;sort=4,0,2' for _ in range(len(soup.find_all('td', class_ = 'left')))]
build_links
```

```python
# soup to find construction links
soup = BeautifulSoup(requests.get('https://rustlabs.com/group=build').content, 'html.parser')
```

```python
construction_links = ['https://rustlabs.com' + soup.find('div', class_ = 'info-block group').find_all('a')[_]['href'] + '#tab=destroyed-by;filter=0,0,1,0,0,0;sort=4,0,2' for _ in range(len(soup.find('div', class_ = 'info-block group').find_all('a')))]
construction_links
```

```python
session = requests.Session()

master = []
for link in build_links:
    soup = BeautifulSoup(session.get(link, timeout = (12, 27)).content, 'html.parser')
    
    for explosive in soup.find_all('tr', {'data-group' : 'explosive'}):
        data_dict = {}
        data_dict['Object']   = soup.find('h1').text
        data_dict['tool']     = explosive.find('td', class_ = 'left padding').text.strip().replace('Semi-Automatic Rifle', '').replace('(right click)', '').replace('Stuck', '').replace('Hunting Bow', '')
        data_dict['quantity'] = int(explosive.find('td', class_ = 'no-padding').text.replace('~', '').replace(',', '').strip())
        data_dict['time']     = explosive.find_all('td')[3].text
        data_dict['type']     = 'Explosive'

        master.append(data_dict)
        
    for explosive in soup.find_all('tr', {'data-group' : 'melee'}):
        data_dict = {}
        data_dict['Object']   = soup.find('h1').text
        data_dict['tool']     = explosive.find('td', class_ = 'left padding').text.strip().replace('Semi-Automatic Rifle', '').replace('(right click)', '').replace('Stuck', '').replace('Hunting Bow', '')
        data_dict['quantity'] = int(explosive.find('td', class_ = 'no-padding').text.replace('~', '').replace(',', '').strip())
        data_dict['time']     = explosive.find_all('td')[3].text
        data_dict['type']     = 'Melee'

        master.append(data_dict)
        
for link in construction_links:
    soup = BeautifulSoup(session.get(link, timeout = (12, 27)).content, 'html.parser')
    
    for explosive in soup.find_all('tr', {'data-group' : 'explosive'}):
        data_dict = {}
        data_dict['Object']   = soup.find('h1').text
        data_dict['tool']     = explosive.find('td', class_ = 'left padding').text.strip().replace('Semi-Automatic Rifle', '').replace('(right click)', '').replace('Stuck', '').replace('Hunting Bow', '')
        data_dict['quantity'] = int(explosive.find('td', class_ = 'no-padding').text.replace('~', '').replace(',', '').strip())
        data_dict['time']     = explosive.find_all('td')[3].text
        data_dict['type']     = 'Explosive'
        master.append(data_dict)
        
    for explosive in soup.find_all('tr', {'data-group' : 'melee'}):
        data_dict = {}
        data_dict['Object']   = soup.find('h1').text
        data_dict['tool']     = explosive.find('td', class_ = 'left padding').text.strip().replace('Semi-Automatic Rifle', '').replace('(right click)', '').replace('Stuck', '').replace('Hunting Bow', '')
        data_dict['quantity'] = int(explosive.find('td', class_ = 'no-padding').text.replace('~', '').replace(',', '').strip())
        data_dict['time']     = explosive.find_all('td')[3].text
        data_dict['type']     = 'Melee'
        master.append(data_dict)
```

```python
import discord
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
import os
from keep_alive import keep_alive

data = pd.read_csv('raiding.csv')

client = discord.Client()

@client.event
async def on_ready():
  print("Ready")

@client.event
async def on_message(message):
    if message.author == client.user:
      return
    
    elif message.content.startswith('!'):
      index = np.argmax([fuzz.token_sort_ratio(data['Object'].unique()[_], str(message.content)) for _ in range(len(data['Object'].unique()))])

      OBJECT = data['Object'].unique()[index]

      stuff_send = data[data['Object'] == OBJECT]

      send_obj = pd.concat([stuff_send[stuff_send['type'] == 'Explosive'], stuff_send[stuff_send['type'] == 'Melee'].sort_values(by = 'quantity').drop_duplicates(subset = ['tool'])[0:3]])

      send_obj = send_obj[(send_obj['tool'] != '40mm HE Grenade') & (send_obj['tool'] != 'High Velocity Rocket')][['Object', 'tool', 'quantity']].reset_index(drop = True)
    
      await message.channel.send(send_obj)
      

client.run(os.getenv('token'))
```
